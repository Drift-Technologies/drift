```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
directory = '../drift-ride-service/data/gtfs_static/'
df = pd.read_csv(directory + 'trips.txt')[['route_id', 'shape_id']]
```

```python
df.head()
```

```python
unique_routes = df.drop_duplicates(subset=['route_id'])
num_routes = len(unique_routes)
num_routes
```

```python
cmap = plt.cm.get_cmap("nipy_spectral", num_routes) 
unique_routes["color"] = [mcolors.to_hex(cmap(i)) for i in range(num_routes)]
```

```python
unique_routes = unique_routes.set_index('route_id')
unique_routes.head()
```

```python
unique_routes.to_csv(directory + 'unique_route_colors.csv')
```

```python
json_data = unique_routes.to_dict()
print(json_data)
```

```python
with open(directory + "unique_routes.json", "w") as f:
    import json
    json.dump(json_data, f, indent=4)
```
